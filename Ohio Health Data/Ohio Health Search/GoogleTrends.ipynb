```python
from pytrends.request import TrendReq
```

```python
# Login to Google. Only need to run this once, the rest of requests will use the same session.
pytrend = TrendReq(hl='en-US', tz=360)

# Create payload and capture API tokens. Only needed for interest_over_time(), interest_by_region() & related_queries()
pytrend.build_payload(kw_list=['health'])

# Interest Over Time
interest_over_time_df = pytrend.interest_over_time()
interest_over_time_df.head()
```

```python
# Interest by Region
interest_by_region_df = pytrend.interest_by_region()
print(interest_by_region_df.head())
```

```python
# Related Queries, returns a dictionary of dataframes
related_queries_dict = pytrend.related_queries()
#print(related_queries_dict)
#pd.DataFrame(related_queries_dict['health'])
#pd.DataFrame.from_dict({(i,j): related_queries_dict[i][j] 
#                           for i in related_queries_dict.keys() 
#                           for j in related_queries_dict[i].keys()},
#                       orient='index')

A=pd.DataFrame(related_queries_dict['health']['rising'])
```

```python
# Get Google Hot Trends data
trending_searches_df = pytrend.trending_searches()
trending_searches_df.head()

```

```python
# Get Google Top Charts
top_charts_df = pytrend.top_charts(cid='politicians', date=201711, cat='business_and_politics')
top_charts_df.head()
```
